```python
morph = np.array(Image.open("morph.png").convert("L"))
morph = (morph <= 128) * 1
plt.imshow(morph, cmap=cm.gray_r)
```

```python
def Circle(r):
    axis = np.linspace(-1, 1, 2 * r + 1)
    x, y= np.meshgrid(axis, axis)
    return (x**2 + y**2 < 1) * 1
plt.matshow(Circle(20), cmap=cm.gray_r)
```

```python
from scipy.ndimage import morphology
plt.imshow(morphology.binary_erosion(morph, Circle(4)), cmap=cm.gray_r)
```

```python
morphEroded = morphology.binary_erosion(morph, Circle(4), iterations=2)
plt.imshow(morphEroded, cmap=cm.gray_r)
```

```python
plt.imshow(morphology.binary_dilation(morphEroded, Circle(6)), cmap=cm.gray_r)
```

```python
plt.imshow(morphology.binary_opening(morph, Circle(5), iterations=2), cmap=cm.gray_r)
```

```python
morph2 = np.array(Image.open("morph2.png").convert("L"))
morph2 = (morph2 <= 128) * 1
plt.imshow(morph2, cmap=cm.gray_r)
```

```python
plt.imshow(morphology.binary_closing(morph2, Circle(5), iterations=2), cmap=cm.gray_r)
```

```python
def mandelbrot(n):
    y, x = np.ogrid[-1.4:1.4:200j, -2.0:0.8:200j]
    z = x + y * 1j
    c = z
    for i in range(n):
        z = z * z + c
        z[z * np.conj(z) > 4] = 2.1
    plt.imshow(np.where(z * np.conj(z) > 4, 1, 0), cmap=cm.gray)
```

```python
mandelbrot(20)
```

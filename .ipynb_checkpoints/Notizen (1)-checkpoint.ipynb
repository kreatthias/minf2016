```python
def countItems(search, array):
    j = 0
    for i in array:
        if search == i:
            j = j+1        
    return j
countItems(42, [1, 2, 3, 42, "42", 100, 42, 3])
```

```python
# Sieb des Eratostenes
def eratos(n):
    a = np.arange(2, n + 1)
    primes = []
    lim = math.sqrt(n)
    p = 2
    while (p <= lim):
        a = a[a % p != 0]
        primes.append(p)
        p = a[0]
    primes.extend(a)
    return primes
```

```python
# Mandelbrotmenge zeichen
# zur Wiederholung: z0 = c
# z1 = z0*z0 + c
# z2 = z1*z1 + c...
# wenn Punkte abhauen, dann weiß
# wenn nicht abgehauen, dann schwarz
# abhauen = Abstand bis zum Nullpunkt größer als 2
```

```python
def mandelbrot(n): # n ist wieoft wir wiederholen bis entschieden ob abhaut
    y,x = np.ogrid[-1.4:1.4:200j, -2.0:0.8:200j] # unsere map, beides Breite von 2,8
    z = x + y * 1j # y wird imaginisiert und dann gekreuzt mit x
    c = z # diese Matrix sind die Startwerte
    for i in range(n):
        z = z*z +c
        z[z * np.conj(z) > 4] = 2.1 # OVERFLOW HACK
    img = np.where(z * np.conj(z) > 4, 1, 0) # z*!z = z² = Abstand von 0
    plt.imshow(img, cmap=cm.gray)
mandelbrot(100)
```

```python
# Stundenübung 7
def blocks(x,y):
    pic = Image.open("abc.png").convert("L")
    w, h = pic.size
    w = (w // x) * x #ganzzahlige Division
    h = (h // y) * y
    pic = pic.crop((0,0,w,h))
    pixels = np.array(pic)
    means = np.zeros((h // y, w // x))
    for i in range(x):
        for j in range(y):
            means += pixels[j::y, i::x]
    means /= (x*y)
    for i in range(x):
        for j in range(y):
            pixels[j::y, i::x] = means
    plt.imshow(pixels, cmap=cm.gray)
blocks(10,10)
```

```python
# gewichteter Mittelwert
# manche Werte sind wichtiger als andere und bekommen einen Faktor, der dann beim Teilen wieder berücksichtigt werden muss
def weightedAvg(a, w):
    s = 0
    for i in range(len(a)):
        s+= a[i] * w[i]
    return s
w = [0.1, 0.4, 0.3, 0.2]
a = [1,2,3,4]
weightedAvg(a,w)
```

```python
def movAbg(a, n):
    r = a.copy()
    for i in range(1,n):
        r += np.concatenate((r[:i] / i, a[:-i]))
    return r / n
plt.plot(w, 'o', movAbg(w, 5), 'o')
        
```

```python
# Faltung
def window(a, weights):    
    a1 = np.full(len(weights)-1, float('NaN')) #x mal den wert NaN auffüllen
    a2 = np.convolve(a, weights, mode='valid') #valid berücksichtigt nur != NaN, full würde mit 0en auffüllen, same ist mittelding
    return np.concatenate((a1,a2))
plt.plot(w, 'o', window(w, [.5, .3333, .1666]), 'o')
```

```python
imageFolder= "images/"
array = []
for file in os.listdir(imageFolder):
    array += [Image.open(imageFolder + file).convert("L")]

    
def getSlice(image, part):
    w, h = image.size
    image = image.crop([1,0,w-1,h])
    w, h = image.size
    teilstueck = int(w/6)
    return image.crop([(part*teilstueck), 0, ((part+1)*teilstueck), h])

black = 1
white = 254

def getNumber(croppedImage):
    array = np.array(croppedImage)
    if (array[28][22] < black):
        return 2
    elif (array[13][7] > white):
        return 0
    elif (array[28][22] > white and array[21][14] > white):
        return 3
    else:
        return 1


def readNumberFromImage(image):
    sum = 0
    for i in range(0,6):
        sum += getNumber(getSlice(image, i))*4**(5-i)
    return sum

def checkSuccess(imageFolder):
    fileCounter = 0
    successCounter = 0
    for file in os.listdir(imageFolder):
        fileCounter += 1
        image = Image.open(imageFolder + file).convert("L")
        if (str(readNumberFromImage(image)) == os.path.basename(file)[:-4]):
            successCounter += 1
    print("Decoded {} of {} files.".format(successCounter, fileCounter))
    
checkSuccess("images/")
```

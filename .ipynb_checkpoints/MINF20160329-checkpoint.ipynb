```python
def blocks(x, y):
    pic = Image.open("wtc.jpg").convert("L")
    w, h = pic.size
    w = (w // x) * x
    h = (h // y) * y
    pic = pic.crop((0, 0, w, h))
    pixels = np.array(pic)
    means = np.zeros((h // y, w // x))
    for i in range(x):
        for j in range(y):
            means += pixels[j::y, i::x]
    means /= (x * y)
    for i in range(x):
        for j in range(y):
            pixels[j::y, i::x] = means
    plt.imshow(pixels, cmap=cm.gray)
```

```python
blocks(1, 1)
```

```python
blocks(20, 20)
```

```python
blocks(5, 50)
```

```python
def weightedAvg(a, w):
    s = 0
    for i in range(len(a)):
        s += a[i] * w[i]
    return s
```

```python
def movAvg(a, n):
    r = a.copy()
    for i in range(1, n):
        r += np.concatenate((r[:i] / i,  a[:-i]))
    return r / n
```

```python
plt.plot(w, 'or', movAvg(w, 5), '*g')
```

```python
plt.plot(w, 'or', movAvg(w, 10), '*g')
```

```python
import pandas as pd
plt.plot(w, 'or', pd.rolling_mean(w, 10), '*g')
```

```python
def window(a, weights):
    return np.concatenate((
        np.full(len(weights) - 1, float('NaN')),
        np.convolve(a, weights, mode='valid')
    ))
```

```python
plt.plot(w, 'or', window(w, [.25, .25, .25, .25]), 'g*')
```

```python
plt.plot(w, 'or', window(w, [.5, .3333, .1666]), 'g*')
```

```python
def GaussKernel(size):
    x = norm.pdf(np.linspace(-2, 2, size))
    y = np.outer(x, x)
    return y / y.sum()
plt.imshow(GaussKernel(30), cmap=cm.gray, interpolation='none')
```

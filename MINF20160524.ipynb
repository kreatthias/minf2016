```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
w = np.zeros(3)
alpha = 0.5
data = np.array([[20, 10, 1], [40, 50, 0]])
```

```python
def reset (w):
    w[:] = 0.0
    
def learn (input, w):
    x = np.hstack((np.ones((len(input), 1)), input[:, 0:2]))
    d = input[:, 2]
    for j in range(len(input)):
        y = 1 if np.sum(w * x[j]) >= 0 else 0
        w[:] = w + alpha * x[j] * (d[j] - y)
    return w
```

```python
learn(data, w)
```

```python
learn(data, w)
```

```python
learn(data, w)
```

```python
def predict(input, w):
    x = np.hstack((np.ones((len(input), 1)), input[:, 0:2]))
    return (np.sum(w * x, axis = 1) >= 0) * 1

predict(data, w)
```

```python
for i in range(10):
    learn(data, w)
predict(data, w)
```

```python
# %load perceptron.py
import numpy as np

def Heaviside (x):
    return 1 if x >= 0 else 0

class Perceptron:
    def __init__ (self, *weights):
        self.weights = np.array(weights)

    def activate (self, *input):
        input = np.hstack(([1], input))
        return Heaviside(np.sum(self.weights * input))
```

```python
p = Perceptron(0, 1, -1)
```

```python
p.activate(2, 1)
```

```python
p.activate(1, 2)
```

```python
p1 = p3 = Perceptron(-1, 1)
p2 = Perceptron(-2, 1, 1)
q = Perceptron(-1, 1, -2, 1)
```

```python
def xor (x, y):
    return q.activate(p1.activate(x), p2.activate(x, y), p3.activate(y))
```

```python
xor(0, 0), xor(0, 1), xor(1, 0), xor(1, 1)
```

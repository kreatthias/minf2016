```python
import numpy as np
import matplotlib.pyplot as plt

pizzaPrices = np.array([[24, 25, 28, 30, 40, 36, 32],
                        [5.5, 8.9, 7.4, 10.9, 19.4, 13.8, 14.0]])
```

```python
X = pizzaPrices[0, :]
Y = pizzaPrices[1, :]
plt.scatter(X, Y, c='r')
```

```python
plt.scatter(X, Y, c='r')
x = np.linspace(20, 45, 50)
y1 = 0.4 * x - 2.8
plt.plot(x, y1)
```

```python
plt.scatter(X, Y, c='r')
x = np.linspace(20, 45, 50)
y2 = 1.1 * x - 23
plt.plot(x, y1, x, y2)
```

```python
XSq = np.sum(X * X)
YSq = np.sum(Y * Y)
XY = np.sum(X * Y)
n = len(X)
```

```python
%matplotlib notebook
```

```python
from mpl_toolkits.mplot3d import Axes3D
a, b = np.meshgrid(np.linspace(-1, 2, 50), np.linspace(-10, 10, 50))
r = XSq * a * a + 2 * np.sum(X) * a * b + n * b * b - 2 * XY * a - 2 * np.sum(Y) * b + YSq

fig = plt.figure()
ax = fig.gca(projection = '3d')
ax.set_xlabel('a')
ax.set_ylabel('b')
ax.plot_surface(a, b, r)
plt.show()
```

```python
def lift(arr):
    arr = np.array(arr)
    return arr.reshape((arr.shape[0], 1))

lift([23, 42, 18])
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(lift(X), lift(Y))
```

```python
model.predict([[35]])
```

```python
plt.figure()
plt.scatter(X, Y, c='r')
plt.plot([20, 45], model.predict([[20], [45]])[:, 0])
```

```python
water = np.array([[5, 7.5, 9, 12, 13.4, 16, 20],
                  [221, 269, 292, 460, 557, 795, 1383]])
X = water[0, :]
Y = water[1, :]

plt.figure()
plt.scatter(X, Y, c='r')
```

```python
model = LinearRegression()
model.fit([[i ** 3] for i in X], lift(Y))
```

```python
x = np.linspace(4, 21, 50)
y = model.predict([[i ** 3] for i in x])

plt.figure()
plt.scatter(X, Y, c='r')
plt.plot(x, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures

featurizer = PolynomialFeatures(degree = 3)
featurizer.fit(lift(X))
featurizer.transform(lift(X))
```

```python
model = LinearRegression()
model.fit(featurizer.transform(lift(X)), lift(Y))

x = np.linspace(4, 21, 50)
y = model.predict(featurizer.transform(lift(x)))

plt.figure()
plt.scatter(X, Y, c = 'r')
plt.plot(x, y)
```
